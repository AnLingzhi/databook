```python
import tushare as ts
import pandas  as pd
import numpy  as np
import matplotlib.pyplot as plt
from   pylab  import  *
```

```python
df = ts.get_hist_data('300036',ktype='M')
```

```python
df
```

```python
def splot(dfx):
    #创建一个画布对象。一个画布可以有多幅子图。
    fig=plt.figure(figsize=(48,4), dpi=80)

    #在画布中创建一个子图，指定一行、一列的第一幅。
    p1=fig.add_subplot(1,1,1)

    #绘制第一条曲线。
    p1.plot(dfx.index,dfx['open'])
    p1.plot(dfx.index,dfx['close'])
    p1.plot(dfx.index,dfx['high'])
    p1.plot(dfx.index,dfx['low'])

    #显示图形。
    plt.show()
```

```python
def splotbar(dfx):
    #创建一个画布对象。一个画布可以有多幅子图。
    fig=plt.figure(figsize=(48,4), dpi=80)

    #在画布中创建一个子图，指定一行、一列的第一幅。
    p1=fig.add_subplot(1,1,1)

    #绘制第一条曲线。
    p1.plot(kind='bar', dfx.index,dfx['p_change'], facecolor='#9999ff', edgecolor='white')
    p1.plot(dfx.index,dfx['p_change'])
    #显示图形。
    plt.show()
```
