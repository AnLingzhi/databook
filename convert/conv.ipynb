```python
from nbconvert import HTMLExporter
```

```python
help(HTMLExporter)
```

```python
def Notebook2Html(path,filename):
    exp = HTMLExporter()
    ostream,res = exp.from_filename(path + filename + ".ipynb")

    file_object = open('../docs/' + filename + '.html', 'w')
    file_object.write(ostream)
    file_object.close()
```

```python
Notebook2Html("../", "quickstart")
```

```python
import os
import os.path
```

```python
def traversefile(workdir):
    count = 1
    print("当前目录：", workdir)
    for filename in os.listdir(workdir):
        file = os.path.join(workdir,filename)
        path,name = os.path.split(file)
        fname,fext = os.path.splitext(os.path.basename(file))
        print("\t文件：", os.path.dirname(file), "," , name,"-", fname,"-",fext)
        if os.path.isdir(file):
            print("子目录：", file)
            count += traversefile(file) #计算该目录下所有的目录个数
        else:
            if fext == ".ipynb":
                Notebook2Html(os.path.dirname(file)+"/", fname)
            continue
    return count
```
