```python
# 绘制价格曲线图。
def splot(dfx):
    fig=plt.figure(figsize=(48,4), dpi=80)
    p1=fig.add_subplot(1,1,1)

    p1.plot(dfx.index,dfx['open'])
    p1.plot(dfx.index,dfx['close'])
    p1.plot(dfx.index,dfx['high'])
    p1.plot(dfx.index,dfx['low'])
    plt.grid(True)
    plt.gcf().autofmt_xdate()
    plt.show()

# 绘制价格变动率和变动的百分比率。    
def splotbar(dfx):
    fig=plt.figure(figsize=(48,4), dpi=80)
    p1=fig.add_subplot(1,1,1)

    p1.plot(dfx.index,dfx['p_change'], color='red')
    p1.bar(dfx.index,dfx['price_change'])
    plt.grid(True)
    plt.gcf().autofmt_xdate()
    plt.show()
```

```python
df2=df.sort_index(ascending=True)
```

```python
splot(df2)
```

```python
splotbar(df2)
```

```python
df2v = pd.DataFrame(df2,columns=("price_change","p_change")).sort_index(ascending=True)
df2c=df2v.cumsum()
df2c.plot(kind='line',figsize=(32,4),grid=True)
df2c.plot(kind='bar',figsize=(32,4),grid=True)
```
